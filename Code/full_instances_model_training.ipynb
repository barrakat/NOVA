### 1.1 Create configuration file (.yaml) to pass for model training

```python
config_data = {
    'path': path_to_tiles,
    'train': path_to_tiles+'/train/images',
    'val': path_to_tiles+'/val/images',
    'test':"/content/drive/MyDrive/content/drive/MyDrive/NOVA_course_deep_learning/data/annotated_data/test/images",
    'names': {
        0: 'tree'
    }
}

with open(path_to_tiles+'/train_config.yaml', 'w') as file:
    yaml.dump(config_data, file)
```
